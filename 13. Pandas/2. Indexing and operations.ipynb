```python
import pandas as pd

s = pd.Series([10, 20, 30, 40, 50], index=["a", "b", "c", "d", "e"])
print(s)
# Access by label
# print(s[["b","c"]])

# Access by position
print("-------")
print(s.iloc[0])
print(s.iloc[1:4])
```

```python
# Operations
# print(s * 2)
print(s[s>25])

# Attributes
print(s.values)
print(s.index)
print(s.dtype)
```
