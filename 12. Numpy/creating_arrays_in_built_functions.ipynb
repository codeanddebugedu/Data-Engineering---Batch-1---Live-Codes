```python
# How to create arrays using inbuilt functions

import numpy as np

zeros=np.zeros((3,6))
print(zeros)
```

```python
ones=np.ones((4,5))
print(ones)
```

```python
arr=np.full((3,3),100)
print(arr)
```

```python
identity=np.eye(5)
print(identity)
```

```python
# Creating array: Ranges & Random
arr=np.arange(0,13,3)
print(arr)
```

```python
arr=np.linspace(1,50,8)
print(arr)
```

```python
arr=np.random.randint(1,100,size=(3,3))
print(arr)
```

```python
arr=np.random.random((3,3))
print(arr)
```
