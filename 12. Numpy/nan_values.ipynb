```python
import numpy as np

arr = np.array([1.0, 2.0, np.nan, 4.0, np.nan, 6.0])


print(np.sum(arr))
print(np.nansum(arr))
print(np.nanmax(arr))
print(np.nanmin(arr))

print(np.isnan(arr))
print(np.count_nonzero(np.isnan(arr)))
```

```python
# Practical Example

temperatures = np.array([22.5, -999, 23.1, 24.0, -999, 22.8, 300, 23.5])

cleaned = np.where(temperatures==-999, np.nan, temperatures)

cleaned=np.where(cleaned>100, np.nan, cleaned)
print(cleaned)

print(np.nanmax(cleaned))
print(np.nanmin(cleaned))
print(np.nanmean(cleaned))
```
