```python
import numpy as np

a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a ** 2)

# Masking
print(a > 20)
```

Broadcasting

```python
# Scalar Broadcast
arr = np.array([1, 2, 3, 4])
print(arr * 10)
print(arr + 10)
```

```python
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])
row = np.array([10, 20, 30])

print(matrix * row)
```
