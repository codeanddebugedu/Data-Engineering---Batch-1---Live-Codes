```python
import pandas as pd
import numpy as np

sales = pd.read_csv("monthly_sales.csv")
sales["date"] = pd.to_datetime(sales["date"])

df = pd.read_csv("employees.csv")
df["city"] = df["city"].str.strip().str.title()
df["department"] = df["department"].str.strip().str.title()
df["salary"] = df["salary"].replace(-999, np.nan)
df = df.drop_duplicates()

result=pd.crosstab(df["department"],df["gender"],margins=True,margins_name="Total")
result
```

```python
result=pd.crosstab(df["department"],df["gender"],normalize="all")*100
result=pd.crosstab(df["department"],df["gender"],normalize="index")*100
result
```

```python
result=pd.crosstab(df["department"],df["gender"],values=df["salary"],aggfunc="max")
result
```

```python
result=pd.crosstab([df["department"],df["designation"]],df["gender"],values=df["salary"],aggfunc="mean")
result
```
