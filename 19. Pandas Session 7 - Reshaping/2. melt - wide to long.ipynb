```python
import pandas as pd
import numpy as np

wide=pd.DataFrame({
 "product": ["Laptop", "Mouse", "Monitor"],
 "Jan_2024": [2500000, 160000, 1400000],
 "Feb_2024": [2400000, 155000, 1420000],
 "Mar_2024": [2600000, 165000, 1450000],
})

# Wide to Long
long=wide.melt(id_vars=["product"],value_vars=["Jan_2024","Feb_2024","Mar_2024"],
               var_name="month",value_name="quantity")
long
```
