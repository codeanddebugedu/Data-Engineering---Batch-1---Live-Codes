```python
import pandas as pd
import numpy as np

sales = pd.read_csv("monthly_sales.csv")
sales["date"] = pd.to_datetime(sales["date"])

summary=sales.pivot_table(
    index=["product","region"],
    columns=sales["date"].dt.year,
    values="revenue",
    aggfunc="sum",
).reset_index()
summary["growth_24_25"]=((summary[2025]-summary[2024])/summary[2024]*100).round(2)
final = summary.melt(
 id_vars=["product", "region", "growth_24_25"],
 var_name="year",
#  value_name="total_revenue"
)
print(final)
```
