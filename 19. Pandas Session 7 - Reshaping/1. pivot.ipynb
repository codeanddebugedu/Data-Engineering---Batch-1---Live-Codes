```python
import pandas as pd
import numpy as np

df = pd.read_csv("employees.csv")
df["city"] = df["city"].str.strip().str.title()
df["department"] = df["department"].str.strip().str.title()
df["salary"] = df["salary"].replace(-999, np.nan)
df = df.drop_duplicates()

# Sales (3 years of monthly data: product x region x month)
sales = pd.read_csv("monthly_sales.csv")
sales["date"] = pd.to_datetime(sales["date"])

sales
```

```python
jan = sales[sales["date"] == "2023-01-01"]
wide_jan_data=jan.pivot(index="product",columns="region",values="quantity")
wide_jan_data
```

pivot_table

```python
result=sales.pivot_table(index="product",columns="region",values=["quantity","revenue"],aggfunc=["sum","mean"])
result
```

```python
result=sales.pivot_table(index="product",columns="region",values="revenue",aggfunc="sum",margins=True,margins_name="Total")
result
```
