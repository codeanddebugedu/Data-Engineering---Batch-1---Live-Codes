```python
import pandas as pd

df = pd.read_csv("employees.csv")
df.isnull().sum()
```

Handling Missing Values: dropna()

```python
# df_clean = df.dropna()
df_clean = df.dropna(subset=["salary","department"])
# df.dropna(how="all") # Very rare, all columns are Missing
df_clean
```

Handling Missing Values: fillna()

```python
df["rating"]= df["rating"].fillna(0)
df["salary"] =df["salary"].fillna(df["salary"].mean())
df["age"] =df["age"].fillna(df["age"].median())
df["department"] =df["department"].fillna(df["department"].mode()[0])
df
```

```python
# Q1: Find the percentage of missing values in the 'rating' column.
df = pd.read_csv("employees.csv")
print(f"Percentage = {df["rating"].isnull().sum() / len(df) * 100:.2f}")
```
