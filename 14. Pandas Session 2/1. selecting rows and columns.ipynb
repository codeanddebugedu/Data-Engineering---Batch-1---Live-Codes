```python
import pandas as pd

df = pd.read_csv("employees.csv")

# Selecting Rows by Position: iloc
df.iloc[0]
df.iloc[0:5]
df.iloc[[0,5,10]]

# Rows and Columns by position
df.iloc[0:3,0:4]
df.iloc[:,0:4]
df.iloc[0:3,:]
df.iloc[-3:]
```

```python
# Selecting rows by Label: loc
df.loc[0]
df.loc[0:5]

# Rows and Columns by label
df.loc[0:2,"name"]
df.loc[0:2,["name","salary"]]
df.loc[0:2,"name":"salary"]
```
