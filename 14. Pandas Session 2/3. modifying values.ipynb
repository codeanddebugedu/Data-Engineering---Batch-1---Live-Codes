```python
import pandas as pd

df = pd.read_csv("employees.csv")
# Modifying Values with loc
df.loc[0,"salary"] = 1000000


df.loc[df["department"] == "HR","salary"] = 50000

# df.loc[df["age"] > 55,"designation"] = "Retired"
# df.loc[df["age"] > 55,"department"] = "Alumni"
df.loc[df["age"] > 55, ["designation", "department"]] = ["Retired", "Alumni"]


# Create new column
df.loc[df["salary"] > 100000,"salary_type"] = "High"
df.loc[df["salary"] < 100000,"salary_type"] = "Standard"
df
```
