```python
import pandas as pd

df = pd.read_csv("employees.csv")

high_salary = df[df["salary"] > 100000]

df[df["department"] == "Engineering"]
df[df["department"] != "HR"]
df[df["age"] >= 40]
```

```python
# Filtering with Multiple Conditions
df[(df["department"] == "Engineering") & (df["salary"] > 100000)]
df[(df["city"] == "Mumbai") | (df["city"] == "Bangalore")]
df[(df["department"] == "Engineering") & (df["salary"] > 100000) & (df["age"] < 45)]
```

```python
# Filtering with isin()
result = df[
    (df["city"] == "Mumbai") |
    (df["city"] == "Delhi") |
    (df["city"] == "Bangalore")
]
result = df[df["city"].isin(["Mumbai","Delhi","Bangalore"])]
result = df[~df["city"].isin(["Mumbai","Delhi","Bangalore"])]
result
```

```python
# Filtering with between()
result = df[(df["age"] >= 25) & (df["age"] <= 35)]
result = df[df["age"].between(25,35)]
result = df[df["age"].between(25,35,inclusive="left")]
result = df[df["age"].between(25,35,inclusive="right")]
result = df[df["age"].between(25,35,inclusive="neither")]
result
```

```python
# Filtering with query()

# Standard Way (verbose way)
result = df[(df["department"] == "Engineering") & (df["salary"] > 80000)]

result = df.query("department == 'Engineering' and salary > 80000")
result = df.query("department != 'HR' and rating >= 4.0")

# What if we get values in a variable
min_salary = 100000
max_age = 40
department="HR"
result = df.query("salary > @min_salary and age < @max_age and department == @department")

result
```

```python
# Filtering String Columns
result = df[df["name"].str.contains("Sharma")] # Case Sensitive
result = df[df["name"].str.contains("sharma" ,case=False)] # Case In-sensitive


result = df[df["name"].str.startswith("A")]
result = df[df["name"].str.len() > 15]
result = df[(df["name"].str.contains("Kumar",case=False)) & (df["department"] == "Engineering")]
result
```
