```python
import pandas as pd
import numpy as np

employees = pd.read_csv("employees.csv")
departments = pd.read_csv("departments.csv")
projects = pd.read_csv("projects.csv")
emp_projects = pd.read_csv("employee_projects.csv")

# Practical example: merge emp_projects with BOTH employees AND projects
step1 = pd.merge(emp_projects,employees[["emp_id","name","department"]],on="emp_id")
final = pd.merge(step1,projects[["project_id","project_name","status"]])
final
```
