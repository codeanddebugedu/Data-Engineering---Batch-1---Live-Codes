```python
import pandas as pd
import numpy as np

employees = pd.read_csv("employees.csv")
departments = pd.read_csv("departments.csv")
projects = pd.read_csv("projects.csv")
emp_projects = pd.read_csv("employee_projects.csv")

# Validate the merge relationship
# "one_to_one": each key appears at most once in BOTH DataFrames
# "one_to_many": each key appears at most once in LEFT
# "many_to_one": each key appears at most once in RIGHT
# "many_to_many": no validation (default)


result=pd.merge(
 employees, departments,
 left_on="department", right_on="dept_name",
 how="left",
 validate="many_to_one")

result
```
