```python
import pandas as pd

df = pd.read_csv("employees.csv")
df["salary"]=df["salary"].fillna(0).astype(int)
df["annual_salary"] = df["salary"] * 12
df["tax"] = df["salary"] * 0.3
df["in_hand_salary"] = df["salary"] - df["tax"]
df[["salary","tax","in_hand_salary","annual_salary"]]
```

```python
import numpy as np

df["is_senior"]=np.where(df['experience_years']>=20,"Yes","No")

conditions=[df["salary"]>=140000,
            df["salary"]>=70000,
            df["salary"]<70000]
choices=["High","Medium","Low"]

df["salary_category"]=np.select(conditions,choices,"Unknown")

df[["experience_years","is_senior","salary","salary_category"]]
```
