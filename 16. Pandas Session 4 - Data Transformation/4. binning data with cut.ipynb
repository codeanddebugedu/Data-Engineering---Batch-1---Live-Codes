```python
import pandas as pd
import numpy as np

df = pd.read_csv("employees.csv")
df['salary']=df['salary'].fillna(0).astype(int)

# Bins are (left, right]
# (0,40000] -> >0 and <=40000
bins=[0,40000, 70000, 100000,150000,300000]
labels=["Very Low","Low","Medium","High","Very High"]

# Bins are [left, right)
df['salary_range']=pd.cut(df['salary'],bins=bins,labels=labels)

df[['salary','salary_range']]
```

```python
# Using qcut
"""
0 43243, 313213,3231

low -> 0
-80000
-160000
- 240000
highest -> 456333

Q1,Q2,Q3,Q4
"""
df['salary_quartile']=pd.qcut(df['salary'],q=4,labels=["Q1","Q2","Q3","Q4"])
df['salary_tercile']=pd.qcut(df['salary'],q=3,labels=["Low","Medium","High"])
df[["salary","salary_quartile","salary_tercile"]]
```
