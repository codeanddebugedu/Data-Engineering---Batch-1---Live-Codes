```python
import pandas as pd
import numpy as np


df = pd.read_csv("employees.csv")
df["salary"]=df["salary"].fillna(0).astype(int)
df["salary_rank"]=df["salary"].rank(ascending=False,method="dense")
new_df=df.sort_values("salary_rank")
new_df[["emp_id","salary","salary_rank"]].head(10)


# Engineering on salary basic rank
df["dept_salary_rank"]=df.groupby("department")["salary"].rank(ascending=False,method="dense")
df[["emp_id","salary","department","dept_salary_rank"]].head(10)
df[df["department"]=="Product"].sort_values("dept_salary_rank")
```
