```python
import pandas as pd
import numpy as np

def salary_category(salary):
    if salary>=120000:
        return "High"
    elif salary >= 70000:
        return "Medium"
    else:
        return "Low"
    
def summary(row):
    return f"Name is {row['name']} and salary is {row['salary']}"


df = pd.read_csv("employees.csv")
df["salary_category"]=df["salary"].apply(salary_category)
df["last_name"]=df["name"].apply(lambda x:x.split()[-1])

df["summary"]=df.apply(summary,axis=1)

df[["salary","salary_category","name","last_name","summary"]]
```

```python
df["bonus_eligible"]=df["ratiNG"].apply(lambda x:"Yes" if x>4.0 else "No")
df[["ratiNG","bonus_eligible"]]


"""
Date parts:

%d → day (01, 15, 31)
%m → month as number (01, 12)
%b → month abbreviated (Jan, Feb)
%B → month full (January, February)
%Y → 4-digit year (2024)
%y → 2-digit year (24)

Time parts:

%H → hour 24hr (00-23)
%I → hour 12hr (01-12)
%M → minute (00-59)
%S → second (00-59)
%p → AM/PM
"""
df["join_date"]=pd.to_datetime(df["join_date"],format="mixed")
df["days_since_joining"] = df["join_date"].apply(lambda x:(pd.Timestamp.now()-x).days)
df["years_since_joining"] = df["join_date"].apply(lambda x:(pd.Timestamp.now()-x).days//365)
df[["join_date","days_since_joining","years_since_joining"]]
```
