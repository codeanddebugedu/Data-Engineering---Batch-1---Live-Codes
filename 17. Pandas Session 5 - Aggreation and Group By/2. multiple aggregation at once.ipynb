```python
import pandas as pd
import numpy as np

df = pd.read_csv("employees.csv")
df["age"]=pd.to_numeric(df["age"])
df["age"]=df["age"].fillna(0).astype(int)

df.groupby("department")["salary"].agg(["mean","sum","max","min","count"])

# salary -> mean, sum
# age -> mean
# rating -> mean, max
# emp_id -> count
df.groupby("department").agg({
    "salary":["mean","sum"],
    "age":["mean"],
    "rating":["mean","max"],
    "emp_id":["count"],
})

```

```python

# salary -> mean
# age -> mean
# rating -> mean
# emp_id -> count
df.groupby("department").agg(
    avg_salary=("salary","mean"),
    sum_salary=("salary","sum"),
    mean_age=("age","mean"),
    avg_rating=("rating","mean"),
    total_emp=("emp_id","count"),
)
```
