```python
import pandas as pd
import numpy as np

df = pd.read_csv("employees.csv")
df["age"]=pd.to_numeric(df["age"])
df["age"]=df["age"].fillna(0).astype(int)

df["dept_avg_salary"]=df.groupby("department")["salary"].transform("mean")
df["dept_vs_avg_salary"]=df["salary"]-df["dept_avg_salary"]
df["dept_salary_rank"]=df.groupby("department")["salary"].transform("rank",ascending=False,method="dense")


df[["emp_id","name","department","salary","dept_avg_salary","dept_vs_avg_salary","dept_salary_rank"]]
```
