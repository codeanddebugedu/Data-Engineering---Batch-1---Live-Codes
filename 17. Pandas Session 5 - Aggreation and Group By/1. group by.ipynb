```python
import pandas as pd
import numpy as np

df = pd.read_csv("employees.csv")
df["age"]=pd.to_numeric(df["age"])
df["age"]=df["age"].fillna(0).astype(int)

print(df.groupby("department")["salary"].sum()) # Total salary
print(df.groupby("department")["salary"].mean()) # Average salary
print(df.groupby("department")["salary"].median()) # Median salary
print(df.groupby("department")["salary"].min()) # Lowest salary
print(df.groupby("department")["salary"].max()) # Highest salary
print(df.groupby("department")["salary"].std()) # Std deviation
print(df.groupby("department")["salary"].count()) # Count (non-null)
print(df.groupby("department").size())
```
